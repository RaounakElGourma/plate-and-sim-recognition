```python
import os
import cv2
import pandas as pd
from ultralytics import YOLO

# === 1. Charger les modÃ¨les ===
plate_model = YOLO("/content/runs/detect/plate_detectiones/weights/best.pt")
char_model  = YOLO("/content/runs/detect/morr_char_detector/weights/best.pt")

# === 2. Dossiers ===
car_images_dir = "/content/drive/MyDrive/projet/data/matricule"  # dossier des images originales
cropped_dir = "/content/drive/MyDrive/projet/cropp"
os.makedirs(cropped_dir, exist_ok=True)

# === 3. Fonction : croper les plaques ===
def crop_plates(img, results, base_name):
    crops = []
    for i, box in enumerate(results[0].boxes.xyxy):
        x1, y1, x2, y2 = map(int, box)
        crop_img = img[y1:y2, x1:x2]
        crop_name = f"{os.path.splitext(base_name)[0]}_plate_{i}.jpg"
        crop_path = os.path.join(cropped_dir, crop_name)
        cv2.imwrite(crop_path, crop_img)
        crops.append(crop_path)
    return crops

# === 4. Fonction : extraire texte depuis une plaque cropÃ©e ===
def extract_text(crop_path):
    img = cv2.imread(crop_path)
    results = char_model(img)
    chars = []
    for box, cls in zip(results[0].boxes.xyxy, results[0].boxes.cls):
        x1, y1, x2, y2 = map(int, box)
        label = results[0].names[int(cls)]
        chars.append((x1, label))
    chars.sort(key=lambda x: x[0])  # tri gaucheâ†’droite
    return "".join([c[1] for c in chars])

# === 5. Boucle principale ===
data = []
for img_name in os.listdir(car_images_dir):
    if img_name.lower().endswith((".jpg", ".jpeg", ".png")):
        img_path = os.path.join(car_images_dir, img_name)
        img = cv2.imread(img_path)

        # DÃ©tection de plaques
        plate_results = plate_model(img, verbose=False)
        cropped_paths = crop_plates(img, plate_results, img_name)

        # Pour chaque plaque dÃ©tectÃ©e â†’ extraire texte
        for crop_path in cropped_paths:
            text = extract_text(crop_path)
            data.append({
                "image_originale": img_name,
                "plaque_crop": os.path.basename(crop_path),
                "texte_extrait": text
            })
            print(f"{crop_path} â†’ {text}")

# === 6. Sauvegarde CSV ===
df = pd.DataFrame(data)
csv_path = "/content/resultat.csv"
df.to_csv(csv_path, index=False)
print(f"\nâœ… RÃ©sultats sauvegardÃ©s dans {csv_path}")

```
